```python
df = pd.read_csv('homeprices.csv')
df
```

```python
plt.scatter(df.area, df.price,color='red',marker='+')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Housing Price Analysis')
```

```python
df.area, df.price
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
reg.fit(new_df,price)
```

```python
reg.predict([[3300]]) # 628715.75342466
```

```python
'''
y = mx + c
y = dependent variable
m = slope
x = independent variable
c = intercept
'''
```

```python
reg.coef_ # 135.78767123
```

```python
reg.intercept_ # 180616.43835616432
```

```python
(135.78767123*3300)+180616.43835616432 # [628715.75342466]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/content/Salary_Data.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
X = df.iloc[:, 0:1].values
y = df.iloc[:, 1].values
```

```python
X, y
```

```python
# Spliting data in train and test data set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.predict([[5.5]]) # 76507.00560722
```
