```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Blurring
image_orig = cv2.imread('/content/fruits.png')
image_orig=cv2.resize(image_orig,(500,500))
```

```python
cv2_imshow(image_orig)
```

```python
# GaussianBlur # kernel values can not be even for height and width (eg: 2,2....4,4...are not taken here) #It gives noise/blur in overall image
Gaussian = cv2.GaussianBlur(image_orig, (5,3),0)
cv2_imshow(Gaussian)
```

```python
median = cv2.medianBlur(image_orig, 3) # boundaries will be gone more if no. increases....works on blurring the edges
cv2_imshow(median)
```

```python
bilateral = cv2.bilateralFilter(image_orig, 9, 75, 75) # targets/blurs small spottings inside the image//works on texture
cv2_imshow(bilateral)
```

```python
# Make border
"""
Syntax: cv2.copyMakeBorder(src, top, bottom, left, right, borderType, value)

Parameters:
src: It is the source image.
top: It is the border width in number of pixels in top direction.
bottom: It is the border width in number of pixels in bottom direction.
left: It is the border width in number of pixels in left direction.
right: It is the border width in number of pixels in right direction.
borderType: It depicts what kind of border to be added. It is defined by flags like cv2.BORDER_CONSTANT, cv2.BORDER_REFLECT, etc dest: It is the destination image
value: It is an optional parameter which depicts color of border if border type is cv2.BORDER_CONSTANT.

"""

image_orig = cv2.imread('/content/fruits.png')
image_orig=cv2.resize(image_orig,(500,500))
image = cv2.copyMakeBorder(image_orig, 2, 10, 10, 10, cv2.BORDER_CONSTANT, None, value = 0)
cv2_imshow(image)
```

```python
image_orig = cv2.imread('/content/fruits.png')
image_orig=cv2.resize(image_orig,(500,500))
image = cv2.copyMakeBorder(image_orig,50, 50, 50, 50, cv2.BORDER_REPLICATE) # it creates that 50,50,50,50 levels of pixel as the margin and replicates the original edge/border as white so we can't see that margin
cv2_imshow(image)
```

```python
src = cv2.imread('/content/fruits.png')
image_orig=cv2.resize(src,(500,500))
image = cv2.cvtColor(image_orig, cv2.COLOR_BGR2GRAY)
cv2_imshow(image)
```

```python
src = cv2.imread('/content/fruits.png')
image_orig=cv2.resize(src,(500,500))
image = cv2.cvtColor(image_orig, cv2.COLOR_BGR2HSV) # HSV might have different range of color....so this is the outcome
cv2_imshow(image)
```
