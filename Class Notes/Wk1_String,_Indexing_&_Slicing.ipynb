```python
#negative indexing & positive slicing
#always add 1 no. for length in -ve index slicing
str1='Peppa prefer Python Programming'
print(str1[-1:-12:-1])
print(str1[-13:-19:-1])
print(str1[-20:-26:-1])
print(str1[-27:-32:-1])
```

```python
#Positive indexing and negative slicing
#start=take positive index, length=check adjacent to -ve slicing, end=give -ve index
str1='Peppa prefer Python Programming'
print(str1[:-12:-1])
print(str1[18:-19:-1])
print(str1[11:-27:-1])
print(str1[4::-1])
```

```python
lst1=['bread','butter','milk','eggs','potatoes','carrots','ginger','garlic','salt','sugar']
print(lst1[::])
print (lst1[0], lst1[1],lst1[2],lst1[3])
print(lst1[4::-1]) # postive index, negative slicing
print(lst1[:-9:-1]) #negative index, negative slicing
print(lst1[6:8:1]) #negative index, positive slicing
print(lst1[-5::-1]) #negative index, negative slicing
```
