```python
import pandas as pd
import matplotlib.pyplot as plt

path = '/content/AB_NYC_2019.csv'
df=pd.read_csv(path)
df
```

```python
df=dfresult
df
```

```python
# 1) Calculate How many properties in each neighbourhood group

count_properties_in_neighbourhood_group = df.groupby('neighbourhood_group')['id'].nunique()
print(count_properties_in_neighbourhood_group )
```

```python
# 2) Which Neighbour Hood Group Has highest prices use bar chart to show this for all neighbourhood groups


group_by_mean = df.groupby('neighbourhood_group')['price'].mean().reset_index()

sorted_groups = group_by_mean.sort_values(by='price', ascending=False)

highest_price_group = sorted_groups.iloc[0]['neighbourhood_group']
highest_price = sorted_groups.iloc[0]['price']

print (sorted_groups)

# Bar chart
plt.figure(figsize=(20, 8))
plt.bar(sorted_groups['neighbourhood_group'], sorted_groups['price'], color='grey')
plt.title('Mean Prices by Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Mean Price')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print(highest_price_group, "neighborhood group has the highest mean price of $", highest_price, ".")
```

```python
# 3) Which Neighbourhood group has lowest prices use bar chart to show this for all neighbourhood groups

group_by_mean = df.groupby('neighbourhood_group')['price'].mean().reset_index()

sorted_groups = group_by_mean.sort_values(by='price', ascending=True)

lowest_price_group = sorted_groups.iloc[0]['neighbourhood_group']
lowest_price = sorted_groups.iloc[0]['price']

print (sorted_groups)

# Bar chart
plt.figure(figsize=(20, 8))
plt.bar(sorted_groups['neighbourhood_group'], sorted_groups['price'], color='skyblue')
plt.title('Mean Prices by Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Mean Price')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print(lowest_price_group, "neighborhood group has the highest mean price of $", lowest_price, ".")
```

```python
# 4) Highest and lowest price airbnb in each Neighbour Hood Group

agg_neighbourhood_group = df.groupby('neighbourhood_group')['price'].agg(['max', 'min'])

agg_neighbourhood_group_in_column = agg_neighbourhood_group.rename(columns={'max': 'Highest_Price', 'min': 'Lowest_Price'})

# Print the result
print(agg_neighbourhood_group_in_column)
```

```python
# 5) Which Host has highest count of properties


# Group by 'host_id' to count the number of properties for each host
host_property_counts = df['host_id'].value_counts()
print (host_property_counts)

# The host with the highest count of properties
highest_property_count_host = host_property_counts.idxmax() #finds the host (host ID) with the maximum count of properties and assigns it to the variable highest_property_count_host
highest_property_count = host_property_counts.max() # returns the maximum value in a Series

print("Host ID", highest_property_count_host, "has the highest count of properties having", highest_property_count, "properties.")
```

```python
# 6) Which neighbourhood is most expensive and cheapest top 10 barchart

neighbourhood_prices = df.groupby('neighbourhood')['price'].mean() # average price for each neighborhood

# Sorting the neighbourhoods by average price in ascending order for cheapest
cheapest_neighbourhoods = neighbourhood_prices.sort_values(ascending=True).head(10)

# Sorting the neighbourhoods by average price in descending order for most expensive
expensive_neighbourhoods = neighbourhood_prices.sort_values(ascending=False).head(10)

# combined DataFrame for plotting
combined_neighbourhoods = pd.concat([cheapest_neighbourhoods, expensive_neighbourhoods])

# bar chart for top 10 cheapest and most expensive neighbourhoods
plt.figure(figsize=(20, 8))
combined_neighbourhoods.plot(kind='bar', color=['green', 'orange'])
plt.title('Top 10 Cheapest and Most Expensive Neighbourhoods')
plt.xlabel('Neighbourhood')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.legend(['Cheapest', 'Most Expensive'])
plt.tight_layout()
plt.show()
```

```python
# 7) Count different type of room_type and do a pie chart showing % of each room type in overall dataset

room_type_counts = df['room_type'].value_counts()
print (room_type_counts)

colors = ['green', 'yellow', 'blue', 'orange']

# pie chart
plt.figure(figsize=(10, 8))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=45, colors=colors)
plt.title('Room Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# 8) scatter plot between longitutude and latitude for different neighbourhood groups

neighbourhood_group_colors = {
    'Bronx': 'blue',
    'Staten Island': 'green',
    'Queens': 'red',
    'Brooklyn': 'orange',
    'Manhattan': 'black',
}

# scatter plots by neighbourhood group
fig, ax = plt.subplots(figsize=(15, 10))

# Loop through each neighborhood group and create a scatter plot
for group, color in neighbourhood_group_colors.items():
    group_data = df[df['neighbourhood_group'] == group]
    ax.scatter(group_data['longitude'], group_data['latitude'], label=group, c=color, alpha=0.5)

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Scatter Plot of Longitude vs. Latitude by Neighborhood Group')
ax.legend()

plt.show()
```

```python
# 9) scatter plot between longitutude and latitude for different room_types

room_type_colors = {
    'Shared room': 'blue',
    'Private room': 'green',
    'Entire home/apt': 'red',
}

fig, ax = plt.subplots(figsize=(15, 10))

for group, color in room_type_colors.items():
    group_data = df[df['room_type'] == group]
    ax.scatter(group_data['longitude'], group_data['latitude'], label=group, c=color, alpha=0.5)

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Scatter Plot of Longitude vs. Latitude by Room Type')
ax.legend()

plt.show()
```

```python
# 10) room_type vs price bar chart for the mean and highest and lowest

#Using agg to calculate  mean, highest, and lowest prices for each room_type
room_type_agg = df.groupby('room_type')['price'].agg(['mean', 'max', 'min'])

plt.figure(figsize=(20, 8))

room_type_agg.plot(kind='bar', colormap='autumn', alpha=0.7)

plt.title('Room Type vs. Price Statistics')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend(['Mean Price', 'Highest Price', 'Lowest Price'])

plt.tight_layout()
plt.show()
```

```python
# 11) Do a similar plot to this


# Create a pivot table to count occurrences of neighbourhood_group within each room_type
pivot_df = df.pivot_table(index='room_type', columns='neighbourhood_group', values='count', aggfunc='count', fill_value=0)

# Create a stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Get the room types (index of pivot_df)
room_types = pivot_df.index

# Get the neighbourhood groups (columns of pivot_df)
neighbourhood_groups = pivot_df.columns

# Initialize a list to store the bottom values for each bar
bottom = [0] * len(neighbourhood_groups)

# Iterate over room types and plot stacked bars
for room_type in room_types:
    values = pivot_df.loc[room_type]
    ax.bar(neighbourhood_groups, values, label=room_type, bottom=bottom)
    bottom = [bottom[i] + values[i] for i in range(len(neighbourhood_groups))]

# Add labels and legend
plt.title('Count of Neighbourhood Groups in Each Room Type')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Count')
plt.legend(title='Room Type')

plt.tight_layout()
plt.show()
```
