```python
# 2) Property with best review_score overall and least score overall

# Defining column names
property_name_column = 'Hotel_Address'
review_score_column = 'Reviewer_Score'

# mean of reviewer score for each property
mean_scores_by_property = df.groupby(property_name_column)[review_score_column].mean().reset_index()

best_property = mean_scores_by_property[mean_scores_by_property[review_score_column] == mean_scores_by_property[review_score_column].max()]
most_positive_property = best_property[property_name_column].values[0]
best_score = best_property[review_score_column].values[0]

least_property = mean_scores_by_property[mean_scores_by_property[review_score_column] == mean_scores_by_property[review_score_column].min()]
least_positive_property = least_property[property_name_column].values[0]
least_score = least_property[review_score_column].values[0]

print('The property with the best review_score overall is', most_positive_property, 'with a score of', best_score)
print('------------------')
print('The property with the least review_score overall is', least_positive_property, 'with a score of', least_score)
```

```python
#4) Which property has highest reviews and lowest  in last 1 month of the dataset maximum date

import pandas as pd
from datetime import datetime, timedelta

df['Review_Date'] = pd.to_datetime(df['Review_Date'])

user_date_input = input("Enter a date (YYYY-MM-DD): ")
user_date = datetime.strptime(user_date_input, '%Y-%m-%d')

# Calculate within one month from the user-provided date
start_date = user_date - pd.DateOffset(months=1)
end_date = user_date

# Filter the dataset to include only the rows within the one-month window from the user's input date
data_within_month = df[(df['Review_Date'] >= start_date) & (df['Review_Date'] <= end_date)]

# Group the data by address and calculate the sum of reviews for each property
property_reviews = data_within_month.groupby('Hotel_Address')['Total_Number_of_Reviews'].sum().reset_index()

# Find the property with the highest and lowest reviews within the one-month window
property_with_highest_reviews = property_reviews[property_reviews['Total_Number_of_Reviews'] == property_reviews['Total_Number_of_Reviews'].max()]
property_with_lowest_reviews = property_reviews[property_reviews['Total_Number_of_Reviews'] == property_reviews['Total_Number_of_Reviews'].min()]

# Print the results
print("Property with the highest reviews within a month from the input date:")
print(property_with_highest_reviews)

print("\nProperty with the lowest reviews within a month from the input date:")
print(property_with_lowest_reviews)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

# Define a function for lemmatization
def lemmatize_text(text):
    words = word_tokenize(text)
    processed_words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(processed_words)

# Define a function for stemming
def stem_text(text):
    words = word_tokenize(text)
    processed_words = [stemmer.stem(word) for word in words]
    return ' '.join(processed_words)

# Apply Lemmatization
df["combined_review_lemmatized"] = df["combined_review"].apply(lemmatize_text)

# Apply Stemming
df["combined_review_stemmed"] = df["combined_review"].apply(stem_text)
```
