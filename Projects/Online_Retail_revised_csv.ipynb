```python
# 2) Create a Dictionary with the 10 countries that have the most 'Quantity' ordered except UK
# Create a bar chart for this

df_no_UK = df[df['Country']!='United Kingdom']

most_qty_ordered = df_no_UK.groupby('Country')['Quantity'].sum().nlargest(10)

sort_country = most_qty_ordered.sort_values(ascending=False)
selected_10_countries_dict = sort_country.to_dict()

print(selected_10_countries_dict)

# Bar chart

Country = list(selected_10_countries_dict.keys())
Quantity = list(selected_10_countries_dict.values())

plt.bar(Country, Quantity, align='center')
plt.xlabel ('Country')
plt.ylabel ('Quantity')
plt.title ('10 Countries with most ordered quantities')
plt.xticks (rotation = 90)
plt.tight_layout()
plt.show()
```

```python
# # 3)  Find Total Sale between two dates entered by user.
# Show a line short for this.

first_date = input("Give first date (MM/DD/YY): ")
second_date = input("Give second date (MM/DD/YY): ")

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

date1 = pd.to_datetime(first_date)
date2 = pd.to_datetime(second_date)

filtered_df = df[(df['InvoiceDate']>=date1) & (df['InvoiceDate']<=date2) & (df['Quantity'] >= 0)]

filtered_df['TotalSale'] = filtered_df['Quantity']*filtered_df['UnitPrice']

total_sales = filtered_df['TotalSale'].sum()

print("Total sales is ", total_sales)

# line chart

sales_by_dates = filtered_df.groupby ('InvoiceDate')['TotalSale'].sum()
print(sales_by_dates)

plt.figure(figsize=(10, 6))
plt.plot(sales_by_dates.index, sales_by_dates.values, linestyle='solid', marker='o', markersize=5)
plt.gcf().autofmt_xdate()

plt.title('Total Sales between Two Dates Entered')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.tight_layout()

plt.show()
```

```python
# 4) Find each Countries Total SALE to find which Countries is highest potential customer.
# Do a bar chart for this.

df['TotalSale'] = df['Quantity']*df['UnitPrice']

total_sales = df.groupby('Country')['TotalSale'].sum()

sort_country = total_sales.sort_values(ascending=False)

selected_countries_dict = sort_country.to_dict()

print(selected_countries_dict)

# bar chart

Country = list(selected_countries_dict.keys())
Sales = list(selected_countries_dict.values())

plt.bar(Country, Sales, align='center')
plt.xlabel ('Country')
plt.ylabel ('TotalSale')
plt.title ('Countries with its total sales')
plt.xticks (rotation = 90)
plt.tight_layout()
plt.show()
```

```python
# 5) Which Customer Id has highest purchase in quantity and total and find out the country of it

df['TotalPurchase'] = df['Quantity']*df['UnitPrice']
select_customer = df.groupby('CustomerID').agg({'Quantity':'sum', 'TotalPurchase':'sum','Country':'first'})

select_customer_sorted = select_customer.sort_values(by=['Quantity', 'TotalPurchase'], ascending=False)

customer_with_highest_quantity = select_customer_sorted.index[0]
customer_with_highest_total_purchase = select_customer_sorted.index[1]

country_with_highest_total_purchase_customer=select_customer_sorted.loc[customer_with_highest_total_purchase, 'Country']

print("Customer ID with the highest quantity is: ", customer_with_highest_quantity)
print ("Customer ID with the highest total purchase is:  ", customer_with_highest_total_purchase)
print("Country of the customer with the highest total purchase is: ", country_with_highest_total_purchase_customer)
```

```python
# 6)Which Customer Id has lowest purchase in quantity and total and find out the country of it

df['TotalPurchase'] = df['Quantity']*df['UnitPrice']
select_customer = df.groupby('CustomerID').agg({'Quantity':sum, 'TotalPurchase':sum,'Country':'first'})

select_customer_sorted = select_customer.sort_values(by=['Quantity', 'TotalPurchase'], ascending=True)


customer_with_lowest_quantity = select_customer_sorted.index[0]
customer_with_lowest_total_purchase = select_customer_sorted.index[1]

country_with_lowest_total_purchase_customer=select_customer_sorted.loc[customer_with_lowest_total_purchase, 'Country']

print("Customer ID with the lowest quantity is: ", customer_with_lowest_quantity)
print ("Customer ID with the lowest total purchase is:  ", customer_with_lowest_total_purchase)
print("Country of the customer with the lowest total purchase is: ", country_with_lowest_total_purchase_customer)
```

```python
# 7) Find the biggest order In Sale (Cost) which country was the order from and find total purchase (Cost) done by that customer ID

df['TotalPurchase'] = df['Quantity']*df['UnitPrice']

biggest_order = df['TotalPurchase'].max()

biggest_order_equals_total_purchase = df[df['TotalPurchase']==biggest_order]

biggest_order_country = biggest_order_equals_total_purchase ['Country'].iloc[0]

customer_ID_biggest_order = biggest_order_equals_total_purchase ['CustomerID'].iloc[0]

total_purchase_by_customer = df[df['CustomerID']==customer_ID_biggest_order]['TotalPurchase'].sum()

print("Country with biggest order: ", biggest_order_country)
print("Customer ID of biggest order: ", customer_ID_biggest_order)
print("Total purchase done by that customer ID: ", total_purchase_by_customer)
```

```python
# 8) Find how many different customers are available for each country save the count in dictionary also return countries with highest and lowest customer count
# do bar chart for customer count. customer count on x axis and countries on y axis

count_customer_by_country = df.groupby('Country')['CustomerID'].nunique().to_dict()

highest_customer_count_by_country = max(count_customer_by_country, key=count_customer_by_country.get)
highest_customer_count = count_customer_by_country[highest_customer_count_by_country]

lowest_customer_count_by_country = min(count_customer_by_country, key=count_customer_by_country.get)
lowest_customer_count = count_customer_by_country[lowest_customer_count_by_country]

print(f"Country with the highest customer count: {highest_customer_count_by_country} ({highest_customer_count} customers)")
print(f"Country with the lowest customer count: {lowest_customer_count_by_country} ({lowest_customer_count} customers)")

# bar chart

plt.figure(figsize=(10, 8))
plt.barh(list(count_customer_by_country.keys()), list(count_customer_by_country.values()))

plt.xlabel ('Count')
plt.ylabel ('Country')
plt.title ('Customer count by countries')
plt.xticks (rotation = 90)
plt.tight_layout()
plt.show()
```

```python
# 9) Find which stock code was ordered the most by each country

ordered_stock_by_country = df.groupby(['Country','StockCode'])['Quantity'].sum()

most_ordered_stock_by_country = ordered_stock_by_country.reset_index()
most_ordered_stock_by_country = most_ordered_stock_by_country.sort_values(['Country','Quantity'], ascending=[True,False])
most_ordered_stock_by_country = most_ordered_stock_by_country.drop_duplicates(subset='Country', keep='first')

print(most_ordered_stock_by_country)
```
